# 2- Analysis

## by payer 

```python
df_payers = df.groupby('payer', as_index=False).agg({'claim_specialty':'count','paid_amount':'sum','month':'min'})
```

```python
df_payers['average_payments_per_service'] = df_payers['paid_amount'] / df_payers['claim_specialty']
```

```python
print(f" Payer with maximum amount paid: {df_payers.nlargest(1, 'paid_amount')['payer'].iloc[0]}")
print(f" Payer with maximum used services: {df_payers.nlargest(1, 'claim_specialty')['payer'].iloc[0]}")
print(f" Payer with highest average payments: {df_payers.nlargest(1, 'average_payments_per_service')['payer'].iloc[0]}")
```

```python
print(f" Payer with minimum amount paid: {df_payers.nsmallest(1, 'paid_amount')['payer'].iloc[0]}")
print(f" Payer with minimum used services: {df_payers.nsmallest(1, 'claim_specialty')['payer'].iloc[0]}")
print(f" Payer with lowest average payments: {df_payers.nsmallest(1, 'average_payments_per_service')['payer'].iloc[0]}")
```

-payer UL is the newest payer of all payers 

```python
top_3_payers = df_payers.nlargest(3, 'paid_amount')
display(top_3_payers)
```

```python
sns.set(style="whitegrid")

columns_to_plot = ['claim_specialty', 'paid_amount', 'average_payments_per_service']
for column in columns_to_plot:
    sorted_df = df_payers.sort_values(by=column, ascending=False)
    plt.figure(figsize=(10, 6))
    sns.barplot(x='payer', y=column, data=sorted_df, palette='viridis')
    plt.title(f'{column} by Payer (Sorted)')
    plt.xlabel('Payer')
    plt.ylabel(column)
    plt.show()
```

## by service category

# 4-Trends

- should filtre the dataframe excluding last 6 months

```python
last_date = pd.to_datetime('2020-01-01').date()
```

```python
df_payment_trend = df.query('month <=@last_date').groupby('month',as_index=False)['paid_amount'].sum()
```

```python
plt.figure(figsize=(12, 8))
sns.lineplot(x='month', y='paid_amount', data=df_payment_trend, label='Actual Data')

coefficients = np.polyfit(range(len(df_payment_trend)), df_payment_trend['paid_amount'], 1)
trend_line = np.poly1d(coefficients)

all_dates = pd.date_range(df_payment_trend['month'].min(), df_payment_trend['month'].max(), freq='M')
plt.plot(all_dates, trend_line(range(len(all_dates))), label='Trend Line', linestyle='dashed', color='red')
future_dates = pd.date_range(df_payment_trend['month'].max() + timedelta(days=1), periods=12, freq='M')
plt.plot(future_dates, trend_line(range(len(all_dates), len(all_dates) + 12)), linestyle='dashed', color='red')

plt.title('Payment Trend with Extended Trend Line')
plt.xlabel('Month')
plt.ylabel('Paid Amount')

plt.legend()
plt.show()
```

- at the end of 2020 should expect a 7.5 millions $ payments monthly on average
